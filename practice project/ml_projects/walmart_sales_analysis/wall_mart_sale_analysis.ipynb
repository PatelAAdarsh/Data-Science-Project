```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
df_features = pd.read_csv('features.csv')
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

# Data Prep 

```python
df_train
```

```python
df_features.columns = df_features.columns.str.lower()
df_train.columns = df_train.columns.str.lower()
df_test.columns = df_test.columns.str.lower()
```

```python
# Convert date column to DateTime type
df_features['date'] = pd.to_datetime(df_features['date'])
df_train['date'] = pd.to_datetime(df_test['date'])
df_test['date'] = pd.to_datetime(df_test['date'])

# Convert DateTime to float
df_features['date'] = pd.to_numeric(df_features['date']).astype(float)
```

# EDA

```python
# column wise summary
df_features.columns
```

```python
# null values
total_features = df_features.isnull().sum(axis = 0).sort_values(ascending = False)
percent_features = ((df_features.isnull().sum(axis = 0)/ df_features.isnull().count(axis=0))*100).sort_values(ascending = False)

# count the number of null values in the column and their perecentage of the total df
missing_data_columns_features = pd.concat([total_features, percent_features], axis = 1, keys=['total', 'percent'])

#training data
total_train = df_train.isnull().sum(axis = 0).sort_values(ascending = False)
percent_train = ((df_train.isnull().sum(axis = 0)/ df_train.isnull().count(axis=0))*100).sort_values(ascending = False)

missing_data_columns_train = pd.concat([total_train, percent_train], axis = 1, keys=['total', 'percent'])

# Testing Data
total_test = df_test.isnull().sum(axis = 0).sort_values(ascending = False)
percent_test = ((df_test.isnull().sum(axis = 0)/ df_test.isnull().count(axis=0))*100).sort_values(ascending = False)

missing_data_columns_test = pd.concat([total_test, percent_test], axis = 1, keys=['total', 'percent'])
```
