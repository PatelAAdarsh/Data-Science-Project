```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_features = pd.read_csv('features.csv')
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_stores = pd.read_csv('stores.csv')

df_stores.columns = df_stores.columns.str.lower()
df_features.columns = df_features.columns.str.lower()
df_train.columns = df_train.columns.str.lower()
df_test.columns =df_test.columns.str.lower()
```

```python
# Dare Preparation
from datetime import datetime

df_sales = df_stores.merge(df_train, how= 'inner', on = 'store')
df_sales['weekly_sales(M)'] = df_sales['weekly_sales']/1_000_000

# date object to datetime
df_sales['date'] = pd.to_datetime(df_sales['date'])

# Extract week number using apply and isocalendar
df_sales['week_n'] = df_sales['date'].apply(lambda x: x.isocalendar()[1])

df_sales['month'] = df_sales['date'].dt.month
df_sales['year'] = df_sales['date'].dt.year

# Disable scientific notation for display
pd.set_option('display.float_format', '{:.3f}'.format)

df_sales
```

### Stores and departments with highest sales for each store type

```python
data = df_sales.groupby(['store','type','dept'])['weekly_sales(M)'].sum().reset_index()

```

```python
filter_data = data[data['dept'].isin([92, 95, 72, 38])].reset_index(drop = True)
```

```python
filter_data
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")



# Create a bar plot
plt.figure(figsize=(25, 6))
sns.barplot(data=filter_data, x='store', y='weekly_sales(M)', hue='dept', palette='viridis')

# Add axis labels and title
plt.xlabel("Store")
plt.ylabel("Weekly Sales (Millions)")
plt.title("Weekly Sales of Each Store by Department")
plt.legend(title='Department', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Set the style
sns.set(style="whitegrid")

# Create a figure and axes for subplots
fig, axes = plt.subplots(4, 1, figsize=(15, 12), sharey=True)
axes = axes.flatten()

# Unique departments
departments = filter_data['dept'].unique()

# Plot for each department
for i, dept in enumerate(departments):
    sns.barplot(data=filter_data[filter_data['dept'] == dept], x='store', y='weekly_sales(M)', hue='type', ax=axes[i], palette='viridis')
    axes[i].set_title(f'Department {dept}')
    axes[i].set_xlabel("Store")
    axes[i].set_ylabel("Weekly Sales (Millions)")
    axes[i].legend(title='Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout
plt.tight_layout()
plt.show()
```

### Identifying Monthly Sales for Each Year

```python
# DataFrame
month_year_sales = df_sales.groupby(['month', 'year'])['weekly_sales(M)'].mean().reset_index()

sns.set(style = "whitegrid")

# Create a bar plot
plt.figure(figsize=(25, 6))
sns.barplot(data=month_year_sales, x='month', y='weekly_sales(M)', hue='year', palette='viridis')

# Add axis labels and title
plt.xlabel("Month", fontsize = 18, fontweight='bold')
plt.ylabel("Weekly Sales (Millions)", fontsize = 18, fontweight='bold')
plt.title("Overall Monthly Sales", fontsize = 22, fontweight='bold')
plt.legend(title='Year', bbox_to_anchor=(1.00, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()


```

### Identifying Week Over Week Sales for Each Year

```python
# DataFrame
week_by_week_sales = df_sales.groupby(['week_n', 'year'])['weekly_sales(M)'].mean().reset_index()
week_by_week_sales
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(18, 10))

sns.lineplot(data=week_by_week_sales, x='week_n', y='weekly_sales(M)',hue='year',
             palette='dark',linewidth = 2.3 )

plt.title(' Week Over Week Sales for Each Year', fontsize = 22, fontweight='bold')
plt.xlabel('Week Number', fontsize = 18, fontweight='bold')
plt.ylabel('Average Sale(M)', fontsize = 18, fontweight='bold')

plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```
